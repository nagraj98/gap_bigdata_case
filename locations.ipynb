```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
import time
import random
```

```python
def extract_locations(baseurl):
    stores_url = baseurl + "/stores"

    resp = requests.get(stores_url)
    html = resp.text

    end = html.find("Desktop schema markup")
    html = html[:end]
    soup = bs(html, 'html.parser')

    rowslist = []
    # state_list = ['Alabama', 'Alaska']
    state_list = ['Alabama', 'Alaska', 'Arizona', 'Arkansas', 'California', 'Colorado', 'Connecticut', 'Delaware', 'District of Columbia', 'Florida', 'Georgia', 'Hawaii', 'Idaho', 'Illinois', 'Indiana', 'Iowa', 'Kansas', 'Kentucky', 'Louisiana', 'Maine', 'Maryland', 'Massachusetts', 'Michigan', 'Minnesota', 'Mississippi']

    
    for state_item in soup.find_all("div", class_="map-list-item is-single"):

        # print(state_list)
        # if len(state_list) == 25:
        #     break

        state = state_item.text.strip()
        
        link = state_item.find('a')
        if link:
            endpoint = link["href"]
            print(state, endpoint)

        

        
        if state not in state_list:
            state_list.append(state)

            # go to the state link and find all the cities
            state_resp = requests.get(baseurl+endpoint)
            state_html = state_resp.text
            state_soup = bs(state_html, 'html.parser')

            # for city_item in state_soup.find("ul", class_="map-list").findChildren() :
            for city_item in state_soup.find_all("div", class_="map-list-item is-single"):

                # link = city_item.find("div", class_="map-list-item is-single").find('a')
                link = city_item.find('a')

                if link:
                    endpoint = link["href"]
                city = link.text.strip()


                # go to city link and get all locations
                time.sleep(max(min(random.gauss(2,1), 3), 1))
                city_resp = requests.get(baseurl+endpoint)
                city_html = city_resp.text
                city_soup = bs(city_html, 'html.parser')

                # for location_item in city_soup.find("ul", class_="map-list").findChildren():
                for location_item in city_soup.find_all("div", class_="map-list-item"):
                    header = location_item.find("div", class_="map-list-item-header")
                    location = header.find("a").text.strip()

                    store_type = header.find(class_="store-type").text.strip()

                    address = location_item.find("div", class_="address").text.strip()
                    address=address.replace("\n", ",")
                    print(city, location, store_type, address)

                    rowslist.append({"State": state, "City": city, "Location": location, "Store Type": store_type, "Address":address, "Endpoint":baseurl+endpoint })
    return rowslist

```

```python
# For banana republic
baseurl = "https://bananarepublic.gap.com"

rowslist = extract_locations(baseurl)
locations_df = pd.DataFrame(rowslist)
locations_df.to_csv("banana_republic_locations.csv")
```
