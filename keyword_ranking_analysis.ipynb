```python
import pandas as pd


df = pd.DataFrame(data)

# List of all unique keywords and brands
unique_keywords = df['keyword'].unique()
unique_brands = df['Brand'].unique()

# Creating an empty DataFrame to store results
result_df = pd.DataFrame(columns=['keyword'] + list(unique_brands))

for keyword in unique_keywords:
    # Initializing a dictionary with keyword and 0 ranks for all brands
    data_dict = {"keyword": keyword}
    data_dict.update({brand: 0 for brand in unique_brands})
    
    # Fetching rows that match the current keyword
    keyword_rows = df[df['keyword'] == keyword]
    
    # Updating the rank for the brands that have a rank
    for _, row in keyword_rows.iterrows():
        brand = row['Brand']
        rank = row['rank']
        data_dict[brand] = rank
    
    # Appending the data dictionary to the result DataFrame
    result_df = result_df.append(data_dict, ignore_index=True)

print(result_df)
```
