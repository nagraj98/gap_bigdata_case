```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
import time
import random

# Helper functions for simulating human behavior

def random_delay(start=1, end=3):
    """Generate a random delay."""
    return time.sleep(random.uniform(start, end))

def faster_typing(element, text):
    """Simulate faster typing."""
    for char in text:
        element.send_keys(char)
        random_delay(0.05, 0.2)

# Main function

def extract_gmap_reviews(driver, search_term):

    # To simulate mouse movements
    actions = ActionChains(driver)

    # Random delay
    random_delay()

    # Faster typing into search box
    search_box = driver.find_element(By.ID, 'searchboxinput')
    faster_typing(search_box, search_term)

    search_box.send_keys(Keys.RETURN)
    random_delay(1, 2)
    search_box.send_keys(Keys.RETURN)
    random_delay(1, 2)



    # Random Scrolling
    for _ in range(random.randint(1, 3)):
        driver.execute_script("window.scrollTo(0, window.scrollY + {})".format(random.randint(50, 300)))
        random_delay(0.5, 1.5)

    review_click = driver.find_elements(By.CLASS_NAME, "LRkQ2")
    random_delay(2, 3)

    review_click[1].click()
    random_delay(1, 2)

    reviews = driver.find_elements(By.CSS_SELECTOR, ".jftiEf")

    extracted_reviews = []

    for review in reviews:

        # Extract review ID
        review_id = review.get_attribute("data-review-id")

        # Extract filled stars count
        filled_stars = review.find_elements(By.CSS_SELECTOR, ".vzX5Ic")
        star_rating = len(filled_stars)

        # Extract review text
        review_text = review.find_element(By.CSS_SELECTOR, ".wiI7pd").text
        
        #Extract months/years ago
        time_ago_element = review.find_element_by_class_name("rsqaWe").text


        extracted_reviews.append({
            'Review_ID': review_id,
            'Review_Text': review_text,
            'Filled_Stars': star_rating,
            'Review_Time': time_ago_element
        })

    # Randomly close and reopen the driver with a probability of 0.3
    if random.random() < 0.3:
        driver.quit()
        random_delay(7, 20)  # Wait a bit before reopening
        driver = webdriver.Firefox()

    return extracted_reviews
```

```python
locations = pd.read_csv("banana_republic_locations.csv")
locations.head()
```

```python
loc_count=161
for idx, row in locations.iterrows():
    try:
        if row["Unnamed: 0"] < 161:
            continue

        search_term = "Banana Republic " + row["Address"]
        
        driver = webdriver.Firefox()
        driver.get("https://www.google.com/maps/")

        reviews = extract_gmap_reviews(driver, search_term)
        df = pd.DataFrame(reviews)
        df.head()

        df.to_csv(f'CSV1/{row["State"]}_{row["City"]}_{row["Location"]}.csv')
        loc_count += 1
        driver.quit()
        
    except Exception as e:
        print(f"Error at index {idx} for search term '{search_term}': {e}")
        driver.quit()
        time.sleep(2)
        
```
