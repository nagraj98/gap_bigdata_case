```python
# Define the ticker symbols for the companies
symbols = ["GPS", "KSS", "M", "AEO", "URBN", "ANF", "DDS"]

# Set the start and end dates
start_date = "2010-04-01"
end_date = "2020-03-31"

# Download historical stock price data for the specified time frame
data = yf.download(symbols, start=start_date, end=end_date)

# Plot the stock prices for all companies to compare them
plt.figure(figsize=(12, 6))

for symbol in symbols:
    plt.plot(data['Adj Close'][symbol], label=symbol)

plt.xlabel("Date")
plt.ylabel("Adj Close Price")
plt.title("Stock Price Comparison")
plt.legend()
plt.grid(True)
plt.show()
```

```python
closing_data = data['Adj Close']
```

```python
# Calculate correlations between the stock prices
correlations = closing_data.corr()

# Display the correlation matrix
print("Correlation Matrix:")
print(correlations)
```

```python
# Define the start and end dates for the first time range
# start_date = "2010-04-01"
cutoff_date = "2015-03-31"
# end_date = "2020-03-31"

# Create subsets of the data for each time range
first_half = closing_data[start_date:cutoff_date]
second_half = closing_data[cutoff_date:end_date]

# Calculate correlations for each time range
correlations_first_half = first_half.corr()
correlations_second_half = second_half.corr()

# Display the correlation matrices for each time range
print("Correlation Matrix (Adj Close Prices) for Time Range 1:")
print(correlations_first_half)

print("\nCorrelation Matrix (Adj Close Prices) for Time Range 2:")
print(correlations_second_half)
```

```python
# Convert the date column to datetime
closing_data["Date"] = pd.to_datetime(closing_data.index)

# Extract the year from the date
closing_data["Year"] = closing_data["Date"].dt.year

# Group by year and calculate the average closing price for each year
average_price_by_year = closing_data.groupby("Year")["Close"].mean()

# Reset the index to have "Year" as a column
average_price_by_year = average_price_by_year.reset_index()
```
